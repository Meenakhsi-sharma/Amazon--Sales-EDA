```python
import seaborn as sns
pivot_table = data.pivot_table(
    values='Amount',
    index='Category',
    columns='Month',
    aggfunc='sum'
)

plt.figure(figsize=(12,6))
sns.heatmap(pivot_table, cmap='coolwarm', annot=True, fmt=".0f")
plt.title('Revenue Heatmap (Month vs Category)')
plt.xlabel('Month')
plt.ylabel('Category')
plt.show()
```

***KEY INSIGHTS FROM REVENUE HEATMAP***

**1. "set" Category is the top revenue driver**
 * The set category shows extremely high revenue compared to all others
 * Peak in April(~15.5M), then slightly declines in May and june
 * **This indicates strong demand and major contribution to total revenue.**

**2. Decline trends after April peak**
 * many category show revenue drop from April and June*
 * set, kurta top shows gradual decline  
      * **possible reasons:**
            * End of seasonal demand* 
            * Reduced promotions* 
            * Market saturations* 

**3. Western Dress shows consistent high performance**
 * high revenue in May (~4.38m) and remains strong in June*
 * **indicates stable demand and growth potetial**
    

**4. Low performance categories**
